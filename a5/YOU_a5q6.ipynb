# A5-Q6: Audio Enhancing

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile
from IPython.display import Audio
from numpy.fft import fft, ifft, fftshift, ifftshift
```

```python
# Load the audio recording
Omega, f = scipy.io.wavfile.read('recording.wav')
Audio(f, rate=Omega)
```

```python
# Some useful values
N = len(f)    # total number of samples
L = N / Omega # length of sound clip (in seconds)
t = np.arange(0,N-1) * L/N  # array of time stamps for samples
```

```python
# Corresponding array of sampled frequencies
omega = np.fft.fftshift(np.arange(-N/2, N/2)) / L
```

```python
plt.plot(f)
plt.title('Sound')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude');
```

## (a)

```python
# === YOUR CODE HERE ===
F = fft(f)
plt.figure(1)
plt.title('Frequency distribution')
plt.xlabel('Freq (hz)')
plt.ylabel('Magnitude');
plt.plot(omega, abs(F))
```

## (b)

```python
# === YOUR CODE HERE ===
threshold = 2400
G = F.copy()
G[abs(omega)>threshold] = 0
F = fft(f)

plt.figure(2)
plt.title('Coefficients filtered')
plt.xlabel('Freq (hz)')
plt.ylabel('Magnitude');
plt.plot(omega, abs(G))
```

## (c)

```python
# === YOUR CODE HERE ===
f0 = ifft(G)
```

```python
# === YOUR CODE HERE ===
plt.plot(f0)
plt.title('Sound')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude');
```
